## To find the number of ratings of highest-rated iPhones on Flipkart India

```python
# Sorting the data based on iphones with highest rating to lowest

highly_rated = apple_sales.sort_values(by=["Star Rating"],ascending=False)
highly_rated
```

```python
#Selecting the top 15 iphones with highest customer rating

highly_rated_15 = highly_rated.head(15)
print(highly_rated_15['Product Name'])
```

```python
#create new DataFrame from existing DataFrame to sort the iphones with their number of customer ratings
highly_rated_rating = highly_rated_15[['Product Name','Number Of Ratings']].copy()
highly_rated_rating
```

```python
# Bar graph showing different iphones with their number of customer ratings

highly_rated_rating.plot(kind='bar', x='Product Name', y='Number Of Ratings', stacked=True,  legend=False, figsize=(12, 7))
plt.title("Number of Ratings of Highest Rated iPhones")
plt.xlabel("Product Name")
plt.ylabel("Number of Ratings")


# According to the bar chart below, iPhone 8 Plus (Gold, 64 GB) has the higest customer rating in flipkart.	
```

```python
#create new DataFrame from existing DataFrame to sort the iphones with their number of customer ratings
highly_rated_reviews = highly_rated_15[['Product Name','Number Of Reviews']].copy()
highly_rated_reviews
```

```python
# Bar graph showing different iphones with their number of customer ratings

highly_rated_reviews.plot(kind='bar', x='Product Name', y='Number Of Reviews', stacked=True,  legend=False, figsize=(12, 7))
plt.title("Number of Reviews of Highest Rated iPhones")
plt.xlabel("Product Name")
plt.ylabel("Number Of Reviews~")


# According to the bar chart below, iPhone 8 Plus (Gold, 64 GB) has the higest number of customer reviews in flipkart.
```

## Relationship between sale price & Number of Customer Ratings of iPhones

```python
figure = px.scatter(data_frame = apple_sales, x="Number Of Ratings",
                    y="Sale Price", size="Discount Percentage", 
                    trendline="ols", 
                    title="Relationship between Sale Price and Number of Ratings of iPhones")
figure.show()


# The figure below shows a negative linear relationship between the sale price of iphones and the number of ratings.
# This shows that the iphones with lower price are sold more in India
```

## Relationship between Discount percentage & Number of Customer Ratings of iPhones

```python
figure = px.scatter(data_frame = apple_sales, x="Number Of Ratings",
                    y="Discount Percentage", size="Sale Price", 
                    trendline="ols", 
                    title="Relationship between Discount Percentage and Number of Ratings of iPhones")
figure.show()

# The graph shows a linear realtionship between the discount percentage on iphones on flipkart and the total number of customer ratings.
# This implies that iPhones with highest discounts are sold more in India.
```
